# 🔬 STLS Scheme Example Using `qupled`
This notebook demonstrates how to solve a STLS scheme using the `qupled` library.

```python
# Install qupled (if available via pip; replace with local path or instructions if needed)
!pip install qupled
```

```python
import qupled.stls as stls
from qupled.output import DataBase
```

## Define and Run the Scheme

```python
# Define the object used to solve the scheme
scheme = stls.Stls()

# Define the input parameters
inputs = stls.Input(10.0, 1.0)
inputs.mixing = 0.5

# Solve scheme
scheme.compute(inputs)
```

## Access and Print Results

```python
# Access the internal energy from the output file
results = DataBase().read_results(scheme.run_id, names=["uint"])
print("Internal energy from the output file: ")
print(results["uint"])

# Compute the internal energy
print("Internal energy from the result class: ")
print(scheme.results.uint)
```
